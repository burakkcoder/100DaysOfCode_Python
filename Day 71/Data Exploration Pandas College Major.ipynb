```python
x = clean_df["Undergraduate Major"][8]
y = clean_df["Mid-Career 90th Percentile Salary"][8]
print(f"{x} has the highest mid-career salary and {x} graduates earn {y}")
clean_df.loc[8]
```

```python
x = clean_df["Undergraduate Major"][49]
y = clean_df["Mid-Career 90th Percentile Salary"][49]
print(f"{x} has the lowest starting median salary and {x} graduates earn {y}")
clean_df.loc[49]
```

```python
x = clean_df["Undergraduate Major"][18]
y = clean_df["Mid-Career 90th Percentile Salary"][18]
print(f"{x} has the lowest mid-career median salary  and {x} graduates earn {y}")
clean_df.loc[18]
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, "Spread", spread_col) # yeni bir column oluşturduk. iki kez yaptığım için aşağıdaki hata çıktı.
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread') #azdan çoğa sıraladık. bu sıra bize en az riskli bölümleri verdi. maaşlar yakın çünkü
low_risk[['Undergraduate Major', 'Spread']].head() #sadece iki column görmek istedik
```

```python
#en yüksek potansiyelli bölümler:
highest_potential = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
#başlangıç ile mezun olduktan sonraki salarylerin farkı. yukarıda spread olarak oluşturmuştuk. bu sefer çoktan aza yaptık
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
clean_df.groupby("Group").count() #group sütunundaki grupları saymış olduk.
```

```python
#gruplara göre ortalama maaşlar:
clean_df.groupby("Group").mean()
```

```python
pd.options.display.float_format = '{:,.2f}'.format #daha güzel yazmak için...
clean_df.groupby("Group").mean()
```
